```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("apartments.csv")
data['price'] = data['price'].str.replace(' ', '').astype(float)
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['price'], axis=1)
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15, 8))
```

```python
numeric_columns = train_data.select_dtypes(include=['float64', 'int64']).columns
numeric_data = train_data[numeric_columns]
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='YlGnBu')
```

```python
train_data['bedrooms'] = np.log(train_data['bedrooms'] + 1)
train_data['bathrooms'] = np.log(train_data['bathrooms'] + 1)
```

```python
train_data.hist(figsize=(15, 8))
```

```python
train_data = train_data.join(pd.get_dummies(train_data.location, dtype=int)).drop(['location'], axis=1)
```

```python
train_data
```

```python
train_data = train_data.join(pd.get_dummies(train_data.rate, dtype=int)).drop(['rate'], axis=1)
```
