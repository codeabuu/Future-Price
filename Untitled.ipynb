```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("apartments.csv")
data['price'] = data['price'].str.replace(' ', '').astype(float)
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['price'], axis=1)
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15, 8))
```

```python
numeric_columns = train_data.select_dtypes(include=['float64', 'int64']).columns
numeric_data = train_data[numeric_columns]
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='YlGnBu')
```

```python
train_data['bedrooms'] = np.log(train_data['bedrooms'] + 1)
train_data['bathrooms'] = np.log(train_data['bathrooms'] + 1)
```

```python
train_data.hist(figsize=(15, 8))
```

```python
train_data = train_data.join(pd.get_dummies(train_data.location, dtype=int)).drop(['location'], axis=1)
```

```python
train_data
```

```python
train_data = train_data.join(pd.get_dummies(train_data.rate, dtype=int)).drop(['rate'], axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap='YlGnBu')
```

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(x="bedrooms", y="bathrooms", data=train_data, hue="price", palette="coolwarm")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train, y_train = train_data.drop(['price'], axis=1), train_data['price']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()
reg.fit(X_train_s, y_train)
```

```python
test_data = X_test.join(y_test)

test_data['bedrooms'] = np.log(test_data['bedrooms'] + 1)
test_data['bathrooms'] = np.log(test_data['bathrooms'] + 1)
test_data = test_data.join(pd.get_dummies(test_data.location, dtype=int)).drop(['location'], axis=1)
test_data = test_data.join(pd.get_dummies(test_data.rate, dtype=int)).drop(['rate'], axis=1)
```

```python
X_test, y_test = test_data.drop(['price'], axis=1), test_data['price']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()
param_grid = {
    "n_estimators": [100, 200, 300],
    "min_samples_split": [2, 4],
    "max_depth": [3, 4, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)
grid_search.fit(X_train_s, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_test_s, y_test)
```
